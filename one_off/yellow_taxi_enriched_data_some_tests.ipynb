```python
from pyspark.sql.functions import sum, col,round,min, max
#which vendor makes the most revenues?
```

```python
df_trips_enriched= spark.read.table("nyctaxi.02_silver.yellow_trips_enriched")
```

```python
df_trips_enriched.agg(min("tpep_pickup_datetime"),max("tpep_pickup_datetime")).display()
```

```python
#which vendor makes the most revenues?
df_trips_enriched.groupBy('vendor').agg(round(sum('total_amount'),2).alias('total_revenue')).orderBy(col('total_revenue').desc()).display()
```

```python
# the most popular borough
df_trips_enriched.groupBy("pu_borough","do_borough").agg(
    sum("passenger_count").alias("total_passengers")
).orderBy(col("total_passengers").desc()).display()
```

```python
#the most common journey (borough to borough)
df_trips_enriched.\
    groupBy("pu_borough","do_borough").\
        agg(sum("passenger_count").alias("total_passengers")
            ).\
    orderBy(col("total_passengers").desc()).display()
```
