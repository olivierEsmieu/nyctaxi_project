```python
from pyspark.sql.functions import col,count,sum,avg,max, min,round
```

```python
df=spark.read.table("nyctaxi.02_silver.yellow_trips_enriched")
```

```python
df2 = df.groupBy(df.tpep_pickup_datetime.cast("date").alias("pickup_date")).\
    agg(count("*").alias("total_trips"),
        round(avg(df.passenger_count),1).alias("avg_passengers_per_trip"),
        round(avg(df.trip_distance),1).alias("avg_distance_per_trip"),
        round(avg(df.fare_amount),2).alias("avg_fare_per_trip"),
        max(df.fare_amount).alias("max_fare"),
        min(df.fare_amount).alias("min_fare"),
        round(sum(df.total_amount),2).alias("total_revenue")
        ).sort(col("pickup_date").asc())
```

```python
df2.write.mode("overwrite").saveAsTable("nyctaxi.03_gold.daily_trip_summary")
```
