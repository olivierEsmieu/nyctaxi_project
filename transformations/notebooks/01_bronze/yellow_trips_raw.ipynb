```python
from pyspark.sql.functions import current_timestamp
from datetime import datetime, date, timezone
from dateutil.relativedelta import relativedelta
```

```python
# 3 mois de moins
three_months_ago = date.today() - relativedelta(months=3)
fomatted_month = three_months_ago.strftime("%Y-%m")
```

```python
df = spark.read.format("parquet").load(f"/Volumes/nyctaxi/00_landing/data_sources/nyctaxi_yellow/{fomatted_month}")
```

```python

df = df.withColumn("processed_timestamp", current_timestamp())
```

```python
df.write.mode("append").saveAsTable("nyctaxi.01_bronze.yellow_trips_raw")
```
